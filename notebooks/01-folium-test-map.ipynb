```python
import folium
from folium.plugins import HeatMap
import pandas as pd
import numpy as np

print("Creating interactive Folium map with realistic methane patterns...")

# Create realistic methane hotspot data (lat, lon, intensity)
methane_hotspots = [
    # Oil/gas regions
    [28.5, 47.5, 0.9],  # Persian Gulf
    [28.0, 50.0, 0.85],  # Iran
    [26.0, 49.0, 0.88],  # Saudi Arabia
    [-5.0, 20.0, 0.7],   # Congo Basin oil
    [4.0, 12.0, 0.75],   # Nigeria
    [37.0, 38.0, 0.8],   # Russia Caucasus
    
    # Agricultural regions
    [15.0, 25.0, 0.65],  # Sub-Saharan Africa (livestock)
    [-15.0, 25.0, 0.7],  # Southern Africa
    [25.0, 78.0, 0.72],  # India (rice + livestock)
    [-10.0, -60.0, 0.68], # Brazil (cattle ranching)
    
    # Industrial/developed (lower but still present)
    [52.0, 8.0, 0.55],   # Germany
    [48.0, 2.0, 0.52],   # France
    [40.0, -75.0, 0.58], # USA East Coast
]

# Create base map (world center)
m = folium.Map(
    location=[20, 0],
    zoom_start=2,
    tiles='OpenStreetMap'
)

# Add heatmap layer
heat_data = [[lat, lon, intensity] for lat, lon, intensity in methane_hotspots]
HeatMap(heat_data, radius=50, blur=25, max_zoom=2, min_opacity=0.3).add_to(m)

# Add LDC countries as markers
ldc_countries = [
    {'name': 'Nigeria', 'lat': 9.0, 'lon': 8.0, 'gdp': 2200},
    {'name': 'Niger', 'lat': 17.6, 'lon': 8.7, 'gdp': 1200},
    {'name': 'Chad', 'lat': 15.5, 'lon': 18.7, 'gdp': 1300},
    {'name': 'Democratic Congo', 'lat': -4.0, 'lon': 21.0, 'gdp': 2400},
    {'name': 'Ethiopia', 'lat': 9.1, 'lon': 40.5, 'gdp': 2500},
    {'name': 'Bangladesh', 'lat': 23.7, 'lon': 90.4, 'gdp': 7600},
]

for country in ldc_countries:
    popup_text = f"<b>{country['name']}</b><br>GDP: ${country['gdp']:,}<br>LDC Status: Yes"
    folium.CircleMarker(
        location=[country['lat'], country['lon']],
        radius=8,
        popup=popup_text,
        color='red',
        fill=True,
        fillColor='red',
        fillOpacity=0.7,
        weight=2
    ).add_to(m)

# Add title
title_html = '''
<div style="position: fixed; top: 10px; left: 50px; width: 350px; 
            background-color: white; border:2px solid grey; z-index:9999; 
            font-size:14px; padding:10px; border-radius:5px;">
    <b>Methane Emissions & Climate Inequality</b><br>
    🔴 Red dots = Least Developed Countries (LDC)<br>
    🔥 Heat map = Methane hotspots (oil/gas, agriculture)<br>
    <br>
    <i>The countries most vulnerable to climate change<br>
    didn't cause the emissions.</i>
</div>
'''
m.get_root().html.add_child(folium.Element(title_html))

# Add layer control
folium.LayerControl().add_to(m)

# Save
m.save('../outputs/01-methane-inequality-folium.html')
print("✓ Interactive map saved: outputs/01-methane-inequality-folium.html")
print("\nOpen it in your browser to explore!")
```
